```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import zscore
```

```python
#use just 4 years of data for now (2015 - 2018)
filenames = [('ucr_leoka_monthly_1960_2020_dta/leoka_monthly_' + str(i) + '.dta') for i in range(2016, 2019)]
filenames
```

```python
leoka_16, leoka_17, leoka_18 = pd.read_stata(filenames[0]), pd.read_stata(filenames[1]), pd.read_stata(filenames[2])
result = pd.concat([leoka_16, leoka_17, leoka_18], ignore_index=True)
```

```python
result
```

```python
result['year'] = result['year'].astype(int)
```

```python
month_map = {'january': 1, 'february': 2, 'march': 3, 'april': 4, 'may': 5, 'june': 6, 'july': 7, 'august': 8, 'september': 9, 'october': 10, 'november': 11, 'december': 12}
result['month_code'] = result['month'].map(month_map) + (result['year'] - 2016)*12
```

```python
result['assaults_total'] = result['assaults_no_injury_total'] + result['assaults_with_injury_total']
```

```python
mask = result.groupby('agency_name')['assaults_total'].transform(lambda x: any(x >= 1))
filtered_result = result[mask]
```

```python
filtered_result.loc[filtered_result['officers_killed_total'] > 0]
```

```python
filtered_result = filtered_result.sort_values(by=['ori', 'month_code'])
filtered_result['post_1'] = 0
pos_treated_rows = filtered_result['officers_killed_total'] > 0
pos_treated_rows.value_counts()
filtered_result.groupby('ori')['post_1'].shift(1)
#filtered_result.loc[pos_treated_rows, 'post_1'] = filtered_result.groupby('ori')['post_1'].shift(1)
#filtered_result['post_1'].value_counts()
#filtered_result = filtered_result.reset_index(drop=True)
```

```python
filtered_result = filtered_result.sort_values(by=['ori', 'month_code'])
filtered_result['post_1'] = 0
filtered_result['post_2'] = 0
filtered_result['post_3'] = 0
filtered_result['post_4'] = 0
filtered_result['post_5'] = 0
filtered_result['post_6'] = 0
filtered_result['post_6+'] = 0

for idx, row in filtered_result.iterrows():
    try:
        next_index, idx_2, idx_3, idx_4, idx_5, idx_6 = idx+1, idx+2, idx+3, idx+4, idx+5, idx+6
        if next_index < len(filtered_result):
            next_row = filtered_result.loc[next_index]
            if row['officers_killed_total'] > 0 and next_row['ori'] == row['ori']:
                filtered_result.at[next_index, 'post_1'] = 1

        idx_2 = idx+2
        if idx_2 < len(filtered_result):
            row_2 = filtered_result.loc[idx_2]
            if row['officers_killed_total'] > 0 and row_2['ori'] == row['ori']:
                filtered_result.at[idx_2, 'post_2'] = 1

        idx_3 = idx+3
        if idx_3 < len(filtered_result):
            row_3 = filtered_result.loc[idx_3]
            if row['officers_killed_total'] > 0 and row_3['ori'] == row['ori']:
                filtered_result.at[idx_3, 'post_3'] = 1

        idx_4 = idx+4
        if idx_4 < len(filtered_result):
            row_4 = filtered_result.loc[idx_4]
            if row['officers_killed_total'] > 0 and row_4['ori'] == row['ori']:
                filtered_result.at[idx_4, 'post_4'] = 1

        idx_5 = idx+5
        if idx_5 < len(filtered_result):
            row_5 = filtered_result.loc[idx_5]
            if row['officers_killed_total'] > 0 and row_5['ori'] == row['ori']:
                filtered_result.at[idx_5, 'post_5'] = 1

        idx_6 = idx+6
        if idx_6 < len(filtered_result):
            row_6 = filtered_result.loc[idx_6]
            if row['officers_killed_total'] > 0 and row_6['ori'] == row['ori']:
                filtered_result.at[idx_6, 'post_6'] = 1
    except:
        continue
    
    #try:
    #    j = idx+7
    #    while filtered_result.loc[j]['ori'] == row['ori']:
    #        filtered_result.at[j, 'post_6+'] = 1
    #        j += 1
    #except:
    #    break
```

```python
#z-scored sizes
filtered_result['force_size_scored'] = zscore(filtered_result['total_employees_officers'])
```

```python
arrest_files = ['ucr_arrests_monthly_index_1974_2018_dta/ucr_arrests_monthly_index_crimes_age_2016.dta', 'ucr_arrests_monthly_index_1974_2018_dta/ucr_arrests_monthly_index_crimes_age_2017.dta', 'ucr_arrests_monthly_index_1974_2018_dta/ucr_arrests_monthly_index_crimes_age_2018.dta']
arrests_16 = pd.read_stata(arrest_files[0])
arrests_17 = pd.read_stata(arrest_files[1])
arrests_18 = pd.read_stata(arrest_files[2])
arrests = pd.concat([arrests_16, arrests_17, arrests_18], ignore_index=True)
```

```python
arrests['total_arrests'] = arrests['theft_tot_arrests'] + arrests['robbery_tot_arrests'] + arrests['rape_tot_arrests'] + arrests['murder_tot_arrests'] + arrests['mtr_veh_theft_tot_arrests'] + arrests['burglary_tot_arrests'] + arrests['arson_tot_arrests'] + arrests['agg_assault_tot_arrests']
arrests['year'] = arrests['year'].astype(int)
```

```python
agg_result = pd.merge(filtered_result, arrests, on=['ori', 'month', 'year'], how='left')
```

```python
agg_result['total_arrests'] = agg_result['total_arrests'].fillna(0)
agg_result['assault_indicator'] = (agg_result['assaults_total'] > 0).astype(int)
```

```python
agg_result = agg_result.drop('post_6+', axis=1)
```

```python
agg_cleaned = agg_result.dropna()
```
